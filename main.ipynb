## **Simple Linear Regression**

This project aims to make a simple model to predict the potential socre of students based on the effort they put in their studies based on the number of hours studied.

**Made by Mohamed Aziz Khezami**


```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
s_data = pd.read_csv("http://bit.ly/w-data")
print("Data imported successfully")

s_data.head(10)
```

```python
plot = s_data.plot(x="Hours", y="Scores", style="o")
plt.title("Hours vs Percentage")
plt.xlabel("Hours Studied")
plt.ylabel("Percentage Score")
plt.show()
```

```python
correlation = s_data.corr()
print(correlation)
s_data.plot(x="Hours", y="Scores", style="o")
plt.title("Hours vs Percentage")
plt.xlabel("Hours Studied")
plt.ylabel("Percentage Score")
plt.show()
```

```python
y = s_data['Scores']
x = s_data.drop('Scores', axis=1)
```

```python
y = pd.DataFrame(y)
y = y.astype(float)

x = pd.DataFrame(x)
x = x.astype(float)
```

```python
from sklearn.model_selection import train_test_split  
from sklearn.preprocessing import StandardScaler
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
reg =LinearRegression()
reg.fit(x_train, y_train)
```

```python
pred = reg.predict(x_test)
pred = pd.DataFrame(pred)
plt.scatter(x_test, y_test, color="blue")
plt.scatter(x_test, pred, color="red")
plt.show()
```

```python
from sklearn.metrics import  r2_score, mean_squared_error, mean_absolute_error
print(' R2 Score :',r2_score(y_test, pred))
print(' MSE :',mean_squared_error(y_test, pred))
print(' MAE :',mean_absolute_error(y_test, pred)) 
```
